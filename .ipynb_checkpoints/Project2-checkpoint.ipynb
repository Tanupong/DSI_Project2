## Create function to calculate Lr.score

```python
def LR_Score(X,y):
    lr = LinearRegression()
    lr.fit(X_scale,y)
    R2_train_score = lr.score(X_scale,y)
    R2_CV_score = cross_val_score(lr,X_scale,y).mean()
    return R2_train_score,R2_CV_score
```

```python
R2_train_score, R2_CV_score = LR_Score(X_scale,y)
print(f"R2 Trainig: {R2_train_score}")
print(f"R2 CV: {R2_CV_score}")
```

A bit overfit? try use lasso

## Create function to get Ridge/Lasso/Elastic Net score

```python
def regularize_score(X,y):
    #Ridge
    ridge = RidgeCV(alphas=np.logspace(0,5,200))
    ridge.fit(X,y)
    ridge_train_r2 = ridge.score(X,y)
    ridge_CV_r2 = cross_val_score(ridge,X,y).mean()
    #Lasso
    lasso = LassoCV(n_alphas=1000)
    lasso.fit(X,y)
    lasso_train_r2 = lasso.score(X,y)
    lasso_CV_r2 = cross_val_score(lasso,X,y).mean()
    #ElasticNet
    en = ElasticNetCV(n_alphas=100,l1_ratio=[0.2,0.4,0.6,0.8])
    en.fit(X,y)
    en_train_r2 = en.score(X,y)
    en_CV_r2 = cross_val_score(en,X,y).mean()
    #Print Result
    print(f"""
    Ridge Training Score: {ridge_train_r2}
    Ridge CV Score: {ridge_CV_r2}
    
    Lasso Training Score: {lasso_train_r2}
    Lasso CV Score:{lasso_CV_r2}
    
    ElasticNet Training Score: {en_train_r2}
    ElasticNet CV Score: {en_CV_r2}
    
    """)
    pass
```

## Model 3: Try  forwarding subset selection

```python
from itertools import combinations
```

```python
all_features = [x for x in house_num_col.columns if x!='saleprice']
len(all_features)
```

```python
result = {}
base_feature_list = []
for i in range(1,len(all_features)):
    CV_score_base = 0
    for feature in [f for f in all_features if f not in base_feature_list]:
        feature_list = [l for l in base_feature_list]
        feature_list.append(feature)
        X = house_num_col[feature_list]
        ss = StandardScaler()
        ss.fit(X)
        X_scale = ss.transform(X)
        train_score,CV_score = LR_Score(X_scale,y)
        if CV_score > CV_score_base:
            CV_score_base = CV_score
            result_dict = {'feature':feature_list,'r2 CV':CV_score}
            result[i] = result_dict
    base_feature_list = result_dict['feature']

result
```

# Model4: Feature engineering with polynomial+sum

```python
# try polynomial on number_kitchen, bsmt_full_bath , totrms_abvgrd , bedroom_abvgr by interaction only
# try combine gr_liv_area+bsmtfin_sf_1+1st_flr_sf+screen_porch+lot_area+wood_deck_sf+enclosed_porch+2nd_flr_sf+low_qual_fin_sf

# Instantiate PolynomialFeatures
poly = PolynomialFeatures(include_bias=False,degree=4,interaction_only=True)
```

```python
poly_columns = ['number_kitchen','bsmt_full_bath','totrms_abvgrd','bedroom_abvgr']
X_before_poly = house_num_col[poly_columns]
X_before_poly
```

```python
X.poly = poly.fit_transform(X_before_poly)
df = pd.DataFrame(X.poly,columns=poly.get_feature_names(poly_columns))
df.head()
```

```python
house_num_col.reset_index(drop=True,inplace=True)
```

```python
for col in df.columns:
    if col not in poly_columns:
        house_num_col[col] = df[col]
house_num_col.shape
```

```python
house_num_col
```

```python
# try combine gr_liv_area + bsmtfin_sf_1 +
#1st_flr_sf + screen_porch + lot_area + wood_deck_sf + enclosed_porch + 2nd_flr_sf + low_qual_fin_sf
house_num_col['total_area_considered'] = house_num_col['gr_liv_area']+house_num_col['bsmtfin_sf_1']+house_num_col['1st_flr_sf']+house_num_col['screen_porch']+ house_num_col['lot_area'] + house_num_col['wood_deck_sf'] + house_num_col['enclosed_porch'] + house_num_col['2nd_flr_sf'] + house_num_col['low_qual_fin_sf']
```
